```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

print("=== AI/ML Engineer Assessment 1: Multi-Agent War Room ===")

# -------------------------------------------------
# INPUTS
# -------------------------------------------------
def create_metrics():
    print("[TRACE] Creating mock dashboard metrics...")
    days = pd.date_range("2026-04-01", periods=10)

    data = pd.DataFrame({
        "date": days,
        "signup_conversion": [25, 26, 27, 28, 24, 23, 21, 20, 18, 17],
        "DAU": [1000, 1020, 1035, 1040, 980, 950, 900, 870, 850, 820],
        "retention_D7": [70, 71, 72, 73, 68, 65, 62, 60, 58, 55],
        "crash_rate": [1, 1, 1.2, 1.1, 2, 2.5, 3, 4, 4.5, 5],
        "latency_p95": [120, 122, 121, 123, 140, 150, 170, 190, 210, 230],
        "support_tickets": [5, 4, 6, 5, 10, 15, 18, 22, 25, 30]
    })
    return data


def create_feedback():
    print("[TRACE] Loading user feedback...")
    return [
        "App is very slow", "Payment failed twice", "Great feature",
        "Amazing UI", "App crashes often", "Very buggy experience",
        "Feature is useful", "Slow response time", "Excellent update",
        "Facing login issue", "Login page freezes", "Payment timeout issue",
        "Feature is helpful", "Loved the design", "Too much lag",
        "Customer support not responding", "Frequent app crash",
        "Nice update overall", "Checkout failed", "App performance degraded",
        "Feature adoption is easy", "Still facing errors",
        "Excellent functionality", "Very poor speed"
    ]


def release_notes():
    print("[TRACE] Loading release notes...")
    return {
        "feature": "New Smart Checkout Feature",
        "known_risks": [
            "High backend API dependency",
            "Payment gateway timeout risk",
            "Possible mobile app crash on low-end devices"
        ]
    }


# -------------------------------------------------
# TOOLS
# -------------------------------------------------
def detect_anomaly(series):
    print("[TOOL] detect_anomaly()")
    mean = np.mean(series)
    std = np.std(series)
    latest = series.iloc[-1]

    return {
        "status": "Anomaly Detected" if latest > mean + std else "Normal",
        "latest_value": float(latest),
        "mean": float(mean),
        "std": float(std)
    }


def sentiment_summary(feedback):
    print("[TOOL] sentiment_summary()")
    positive_words = ["great", "excellent", "amazing", "useful", "nice", "helpful", "loved"]
    negative_words = ["slow", "failed", "crash", "buggy", "issue", "lag", "error", "poor"]

    pos = 0
    neg = 0

    for text in feedback:
        text = text.lower()
        if any(word in text for word in positive_words):
            pos += 1
        if any(word in text for word in negative_words):
            neg += 1

    return {"positive": pos, "negative": neg}


def trend_comparison(series):
    print("[TOOL] trend_comparison()")
    first = series.iloc[0]
    last = series.iloc[-1]

    return {
        "start_value": float(first),
        "end_value": float(last),
        "trend": "Decreasing" if last < first else "Increasing"
    }


# -------------------------------------------------
# AGENTS
# -------------------------------------------------
class ProductManagerAgent:
    def analyze(self, metrics):
        print("[AGENT] Product Manager")
        adoption_trend = trend_comparison(metrics["signup_conversion"])

        return {
            "summary": "User adoption is falling" if adoption_trend["end_value"] < 20 else "Adoption stable",
            "metric_reference": adoption_trend
        }


class DataAnalystAgent:
    def analyze(self, metrics):
        print("[AGENT] Data Analyst")
        return {
            "crash_analysis": detect_anomaly(metrics["crash_rate"]),
            "latency_analysis": detect_anomaly(metrics["latency_p95"]),
            "ticket_trend": trend_comparison(metrics["support_tickets"])
        }


class MarketingAgent:
    def analyze(self, feedback):
        print("[AGENT] Marketing")
        return sentiment_summary(feedback)


class RiskCriticAgent:
    def analyze(self, metric_report, sentiment, notes):
        print("[AGENT] Risk Critic")
        risks = []

        if metric_report["crash_analysis"]["status"] == "Anomaly Detected":
            risks.append("High crash risk")

        if metric_report["latency_analysis"]["status"] == "Anomaly Detected":
            risks.append("High latency risk")

        if sentiment["negative"] > sentiment["positive"]:
            risks.append("Negative customer sentiment")

        risks.extend(notes["known_risks"])
        return risks


class CoordinatorAgent:
    def final_decision(self, risks):
        print("[AGENT] Coordinator")
        if len(risks) >= 3:
            return "ROLL BACK"
        elif len(risks) == 2:
            return "PAUSE"
        return "PROCEED"


# -------------------------------------------------
# WORKFLOW
# -------------------------------------------------
print("[TRACE] Starting workflow...\n")

metrics = create_metrics()
feedback = create_feedback()
notes = release_notes()

pm = ProductManagerAgent()
data_agent = DataAnalystAgent()
marketing = MarketingAgent()
risk = RiskCriticAgent()
coord = CoordinatorAgent()

pm_report = pm.analyze(metrics)
data_report = data_agent.analyze(metrics)
marketing_report = marketing.analyze(feedback)
risks = risk.analyze(data_report, marketing_report, notes)
decision = coord.final_decision(risks)

print("\n[TRACE] Generating final output...\n")

output = {
    "decision": decision,
    "rationale": {
        "pm": pm_report,
        "data": data_report,
        "marketing": marketing_report
    },
    "risk_register": risks,
    "action_plan_24_48_hours": [
        {"action": "Fix crash issue", "owner": "Engineering Team", "timeline": "24 hours"},
        {"action": "Improve API latency", "owner": "Backend Team", "timeline": "24 hours"},
        {"action": "Notify customers", "owner": "Marketing Team", "timeline": "48 hours"}
    ],
    "communication_plan": {
        "internal": "Escalate issue to engineering, product and leadership",
        "external": "Temporary rollback due to elevated crash and latency issues"
    },
    "confidence_score": 0.97,
    "confidence_improves_if": [
        "Crash rate remains below 2% for next 24 hours",
        "p95 latency remains below 150 ms",
        "Negative feedback drops below positive feedback",
        "Support tickets fall below 10"
    ],
    "release_notes": notes,
    "generated_at": str(datetime.now())
}

print(json.dumps(output, indent=4))

with open("output.json", "w") as f:
    json.dump(output, f, indent=4)

print("\n[SUCCESS] output.json saved successfully")
```
